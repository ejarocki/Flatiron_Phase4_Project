### Feature Engineering: Return on Investment

In order to calculate the historical ROI values, we used the average price of 2009-01~2009-12 for the initial price and the average of 2017-05!2018-04 for final price. The formula for ROI is as follows:

```python
![roi](time-series/photos/roi_formula.png)
```

```python
#ROI Calculation
df['ROI'] = ((df.iloc[:,260:272].mean(axis=1))/(df.iloc[:,160:172].mean(axis=1))-1)*100 
```

```python
df = df.sort_values(['ROI'], ascending=[False])
df[:10] #Descending ROI Order
```

Here we want to focus on the prices that have a coefficient of variance that is associated with a lower amount of risk.  We selected zipcodes according to a coefficient of variation below the 60th percentile, meaning that these zipcodes are stable and reliable markets.

```python
df['std']=df.loc[:,'2009-01':'2018-04'].std(skipna=True, axis=1)
#Calculate historical mean value
df['mean']=df.loc[:,'2009-01':'2018-04'].mean(skipna=True, axis=1)
#Calculate coefficient of variance
df['CV']=df['std']/df['mean']
#Define upper limit of CV according to risk profile.
upper_cv = df.CV.quantile(.6)
best_ROI = df[df['CV']<upper_cv].sort_values('ROI',axis=0,ascending=False)[:10]
best_ROI
```

# Reshape from Wide to Long Format

After selecting the zipcodes with the highest ROI in each region, we reshaped our data into a long format, while also applying `to_datetime` to change our dataframe into a date-time series.

```python
def melt_data(df):
    """
    Takes the zillow_data dataset in wide form or a subset of the zillow_dataset.  
    Returns a long-form datetime dataframe 
    with the datetime column names as the index and the values as the 'values' column.
    
    If more than one row is passes in the wide-form dataset, the values column
    will be the mean of the values from the datetime columns in all of the rows.
    """
    
    melted = pd.melt(df, id_vars=['RegionName', 'RegionID', 'SizeRank', 'City', 'State', 'Metro', 'CountyName','ROI','std','mean','CV','USregion'], var_name='time')
    melted['time'] = pd.to_datetime(melted['time'], infer_datetime_format=True)
    melted = melted.dropna(subset=['value'])
    return melted.groupby('time').aggregate({'value':'mean'})
```

# Testing for Stationarity

### To make our top 5 zipcodes stationary time series, we used differencing

```python
#differencing our time series data
a_diff = a_ts.diff().dropna()  
b_diff = b_ts.diff().dropna()
c_diff = c_ts.diff().dropna()
d_diff = d_ts.diff().dropna()
e_diff = e_ts.diff().dropna()
```

By utilizing the Dickey-Fuller Test, we determined whether the time series data for each of our zipcodes were stationary. The relevant metric we looked at was the p-value. If the p-value was below 0.05, then the data is stationary, but if the p-value was above 0.05, then the data is nonsatationary.

```python
from statsmodels.tsa.stattools import adfuller
#A time series Differencing
result = adfuller(a_diff)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
print('Reject the null hypothesis. Data is Stationary')
```

```python
#B time series differencing 
from statsmodels.tsa.stattools import adfuller
#from index 160
result = adfuller(b_diff)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
print('Reject the Null hypothesis. Data is Stationary')
```

```python
from statsmodels.tsa.stattools import adfuller
#c time series differencing
#from index 160
result = adfuller(c_diff)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
print('Reject the Null Hypothesis. Data is Stationary')
```

```python
from statsmodels.tsa.stattools import adfuller
#from index 160
result = adfuller(d_diff)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
print('Fail to Reject the Null. Data is not Stationary')
```

```python
from statsmodels.tsa.stattools import adfuller
#from index 160
result = adfuller(e_diff)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
print('Fail to Reject the Null. Data is not Stationary')
```

### We can see that A,B, & C were stationary while D & E are not

For our train test split for all five zipcode data, we placed the cutoff point at `'2016-12-01'`, the period prior to this cutoff point being the train and the period after the cutoff point being the test.

```python
#2016 train test split 2017-2018 test
cutoff = "2016-12-01"
cutoff
```

```python
#train-test split for all 5 time series
train_a = a_ts['2009':cutoff]
test_a = a_ts[cutoff:]
train_b = b_ts['2009':cutoff]
test_b = b_ts[cutoff:]
train_c = c_ts['2009':cutoff]
test_c = c_ts[cutoff:]
train_d = d_ts['2009':cutoff]
test_d = d_ts[cutoff:]
train_e = e_ts['2009':cutoff]
test_e = e_ts[cutoff:]
```

# Model A: Monroe, PA (Northeast)

Before we conduct our ARIMA model, we first made a naive baseline model, which was a shift forward. By comparing the naive baseline model with the train, we calculated the error term, root mean squared error, as a point of comparison.

### Baseline Model

```python
# we can perform this with the shift operator
# The prediction for the next day is the original series shifted to the future by one day.
naive_a = train_a.shift(1)
naive_a
```

```python
#This visualizations shows how the shift is applied to the original dataset
fig, ax = plt.subplots()

train_a[:].plot(ax=ax, c='r', label='original')
naive_a[:].plot(ax=ax, c='b', label='shifted')
ax.set_title('naive')
ax.legend(['original','shifted']);
```

### Exploring ACF and PCF

Exploring the ACF and PCF graphs can help select the MA and AR terms for the ARIMA model.

Generally, we use an autocorrelation plot to predict MA terms. It essentially models what are sometimes called "impulses" or "shocks" whose effect accounts for the propogation of correlation from one lag to the other.  Our differenced series shows a positive significant correlation at a lag of 1, which suggests adding 1 MA term.

```python
plot_acf(train_a.diff().dropna())
```

On the other hand, the partial autocorrelation function can be used to get an idea of a potential AR term.  Our differenced series shows significant correlation at a lag of 2, which suggests adding 2 to the AR term.

```python
plot_pacf(train_a.diff().dropna())
```

### First ARIMA model

Since the ACF and the PACF graphs indicate that 1 lag for the MA and 2 lags for the AR terms will create the optimal model with the lowest AIC and RSME values, let's try it out. We must not forget that we differenced earlier to remove stationarity, thus we must also include a differencing value of 1 in our ARIMA model.

```python
#fitting train into ARIMA model
arima_a = ARIMA(train_a, order=(2, 1, 1)).fit()
y_hat_a = arima_a.predict(btyp='levels')
```

```python
fig, ax = plt.subplots()
ax.plot(y_hat_a)
ax.plot(train_a);
```

```python
arima_a.summary()
```

Observing the SARIMAX results, we can see a AIC value of 1570. Lower the AIC value, the better the model.

### Testing

Now, the model will be applied to the test data.

```python
y_hat_test_a = arima_a.predict(start=test_a.index[0], end=test_a.index[-1],typ='levels')

fig, ax = plt.subplots()
ax.plot(y_hat_test_a);
```

Comparing the actual test prices with the predicted prices made under the ARIMA model, we can gauge the effectiveness of our model. Finding the RSME of the test data will offer a more concrete metric for comparison.

```python
fig, ax = plt.subplots()
ax.plot(y_hat_test_a)
ax.plot(test_a);
```

### We will move on to a SARIMA model which takes seasonality into account

```python
p = q = range(0, 2)
pdq = list(itertools.product(p, [1], q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, [1], q))]
print('Examples of parameter for SARIMA...')
for i in pdq:
    for s in seasonal_pdq:
        print('SARIMAX: {} x {}'.format(i, s))
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod=SARIMAX(train_c,
                         order=param,
                         seasonal_order=param_seasonal,
                         enforce_stationarity=False,
                         enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{} - AIC:{}'.format(param,param_seasonal,results.aic))
        except: 
            print('Oops!')
            continue
```

### We can see that the best SARIMA values are ARIMA(1, 1, 1)x(1, 1, 1, 12) - with an AIC of 1156.3232504590499

Let's see how the SARIMA model does making predictions of the test set

```python
sari_mod_c = SARIMAX(train_c,
                  order=(1, 1, 1),
                  seasonal_order=(1, 1, 1, 12),
                  enforce_stationarity=False,
                  enforce_invertibility=False).fit()
```

```python
sari_test_c = sari_mod_c.predict(start=test_c.index[0], end=test_c.index[-1],typ='levels')

from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(test_c,sari_test_c))
```

# SARIMA Model for Region E (Benton City, WA)

Since Benton City, WA had unstationary data, we decided to use a SARIMA model, which is more effective in addressing seasonality.

```python
#Train Test Split
train_e = e_ts['2009':cutoff]
test_e = e_ts[cutoff:]
```

We created a function that iterates a list of SARIMA combinations with different parameters. These parameters include the p,d,q as well as the seasonal p,d,q. Subsequently, we shall observe which of these combination yields the lowest AIC value and use that particular set of parameters as our model.

```python
p = q = range(0, 2)
pdq = list(itertools.product(p, [1], q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, [1], q))]
print('Examples of parameter for SARIMA...')
for i in pdq:
    for s in seasonal_pdq:
        print('SARIMAX: {} x {}'.format(i, s))
```

```python
#Using this for loop to see how to optimize hyper paramaters
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod=SARIMAX(train_e,
                         order=param,
                         seasonal_order=param_seasonal,
                         enforce_stationarity=False,
                         enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{} - AIC:{}'.format(param,param_seasonal,results.aic))
        except: 
            print('Oops!')
            continue
```

```python
#best version of the above hypertuning to get the lowest AIC value
#the order accounts for 1 AR term, 1 Differenced term, and 1 MA term. 
#the seasonal order is also optimized in the below order
sari_mod_e =SARIMAX(train_e,
                  order=(1, 1, 1),
                  seasonal_order=(1, 1, 1, 12),
                  enforce_stationarity=False,
                  enforce_invertibility=False).fit()
```

```python
y_hat_train = sari_mod_e.predict(typ='levels')
y_hat_test = sari_mod_e.predict(start=test_e.index[0], end=test_e.index[-1],typ='levels')

fig, ax = plt.subplots()
ax.plot(train_e, label='train')

ax.plot(y_hat_train, label='train_pred')
ax.plot(y_hat_test, label='test_pred')

plt.legend();
```

```python
#our model is off by 12964 for pricing as represented by our metric of root mean squared error
np.sqrt(mean_squared_error(test_e, y_hat_test))
```
